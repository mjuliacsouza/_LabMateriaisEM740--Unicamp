```python
import pandas as pd
import seaborn as srn
import statistics  as sts
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
liga2024EN = pd.read_csv("liga2024EN.csv",sep=";")
liga2024EA = pd.read_csv("liga2024EA.csv",sep=";")
liga6351EN = pd.read_csv("liga6351EN.csv",sep=";")
liga6351EA = pd.read_csv("liga6351EA.csv",sep=";")
liga6351EA
```

```python
plt.figure(1)
plt.subplot(2,2,1)
plt.scatter(x=liga2024EN.iloc[:,1],y=liga2024EN.iloc[:,0])
plt.ylabel("Dureza RB")
plt.title("Envelhecimento Natural")
plt.legend(["Liga 2024"])
plt.subplot(2,2,2)
plt.scatter(x=liga2024EA.iloc[:,1],y=liga2024EA.iloc[:,0],color='red')
plt.title("Envelhecimento Artificial")
plt.legend(["Liga 2024"])
plt.subplot(2,2,3)
plt.scatter(x=liga6351EN.iloc[:,1],y=liga6351EN.iloc[:,0],color='green')
plt.xlabel("Tempo em minutos")
plt.ylabel("Dureza RB")
plt.legend(["Liga 6351"])
plt.subplot(2,2,4)
plt.scatter(x=liga6351EA.iloc[:,1],y=liga6351EA.iloc[:,0],color='yellow')
plt.xlabel("Tempo em horas")
plt.legend(["Liga 6351"])
```

```python
#Ajuste de curvas em polin√¥mio de grau 1,2 e 4
plt.figure(1)
plt.subplot(2,2,1)
plt.scatter(x=liga2024EN.iloc[:,1],y=liga2024EN.iloc[:,0])
a,b,c = np.polyfit(liga2024EN.iloc[:,1],liga2024EN.iloc[:,0],2)
t = np.linspace(0,170,100)
plt.plot(t,a*t**2+b*t+c)
plt.subplot(2,2,2)
plt.scatter(x=liga2024EA.iloc[:,1],y=liga2024EA.iloc[:,0])
a,b = np.polyfit(liga2024EA.iloc[:,1],liga2024EA.iloc[:,0],1)
t = np.linspace(0,90,100)
plt.plot(t,a*t+b)
plt.subplot(2,2,3)
plt.scatter(x=liga6351EN.iloc[:,1],y=liga6351EN.iloc[:,0])
a,b,c = np.polyfit(liga6351EN.iloc[:,1],liga6351EN.iloc[:,0],2)
t = np.linspace(0,170,100)
plt.plot(t,a*t**2+b*t+c)
plt.subplot(2,2,4)
plt.scatter(x=liga6351EA.iloc[:,1],y=liga6351EA.iloc[:,0])
a,b,c = np.polyfit(liga6351EA.iloc[:,1],liga6351EA.iloc[:,0],2)
t = np.linspace(0,90,100)
plt.plot(t,a*t**2+b*t+c)
```
